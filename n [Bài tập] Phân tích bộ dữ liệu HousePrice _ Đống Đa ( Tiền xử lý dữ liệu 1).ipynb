### 4. Lọc thông tin những bất động sản ở trong ngõ thành bộ dữ liệu nhà ngõ

```python
## Lọc bộ dữ liệu nhà trong Ngõ
df_ngo = df2[df2['title'].str.find("ngõ") != -1]
df_ngo
```

### 5. Tính toán giá/m2 (đơn vị triệu/m2) với loại hình nhà ngõ

```python
df_ngo['gia/m2'] = df_ngo['price']/df_ngo['area']
df_ngo
```

### 6. Phát hiện giá trị ngoại lai của các thuộc tính: diện tích, giá/m2 bằng phương pháp IQR

```python
### area
Q1_a = df_ngo['area'].quantile(0.25)
Q3_a = df_ngo['area'].quantile(0.75)
IQR_a = Q3_a - Q1_a
GHD_a = Q1_a - 1.5* IQR_a
GHT_a = Q3_a + 1.5 * IQR_a
print('Giới hạn trên', GHT_a)
print('Giới hạn dưới',GHD_a)

## phát hiện giá trị ngoại lai bằng những điểm lớn hơn giới hạn trên hoặc nhỏ hơn GHD
df_ngo[(df_ngo['area']>GHD_a)|(df_ngo['area']<GHT_a)]
```

```python
## Có area âm, cần loại bỏ
```

```python
df_ngo1 = df_ngo[df_ngo['area']>0]
df_ngo1
```

```python
## Tính lại GHT, GHD
### area
Q1_a1 = df_ngo1['area'].quantile(0.25)
Q3_a1 = df_ngo1['area'].quantile(0.75)
IQR_a1 = Q3_a1 - Q1_a1

GHD_a1 = Q1_a1 - 1.5* IQR_a1
GHT_a1 = Q3_a1 + 1.5* IQR_a1

print('Q1: ', Q1_a1, ', Q3: ' , Q3_a1)
print('Giới hạn trên', GHT_a1)
print('Giới hạn dưới',GHD_a1)

## phát hiện giá trị ngoại lai bằng những điểm lớn hơn giới hạn trên hoặc nhỏ hơn GHD
df_ngo1[(df_ngo1['area']>GHT_a1)|(df_ngo1['area']<GHD_a1)]
```

```python
## loại bỏ giá/m2 âm
df_ngo2 = df_ngo1[df_ngo1['gia/m2']>0]
df_ngo2
```

```python
## Tính GHT, GHD
### gia/m2
Q1_g = df_ngo2['gia/m2'].quantile(0.25)
Q3_g = df_ngo2['gia/m2'].quantile(0.75)
IQR_g = Q3_g - Q1_g

GHD_g = Q1_g - 1.5* IQR_g
GHT_g = Q3_g + 1.5* IQR_g

print('Q1: ', Q1_g, ', Q3: ' , Q3_g)
print('Giới hạn trên', GHT_g)
print('Giới hạn dưới',GHD_g)

## phát hiện giá trị ngoại lai bằng những điểm lớn hơn giới hạn trên hoặc nhỏ hơn GHD
df_ngo2[(df_ngo2['gia/m2']>GHT_g)|(df_ngo2['gia/m2']<GHD_g)]
```
